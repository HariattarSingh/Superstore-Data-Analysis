2. Exploratory Data Analysis

1. Basic Metrics :
What are the total sales and total profit for the Superstore?

```python
total_sales = df['Total Sales'].sum()
total_profit = df['Profit'].sum()

print(f"Total Sales: {total_sales:}")
print(f"Total Profit: {total_profit:}")
```

What is the average order value?

```python
total_sales = df['Total Sales'].sum()
unique_orders = df['Order ID'].nunique()

average_order_value = total_sales / unique_orders
print(f"Average Order Value: {average_order_value:}")
```

2. Product Analysis: Which product has the highest sales? Which has the highest profit?

```python
product_sales = df.groupby('Product Name')['Total Sales'].sum()
product_profit = df.groupby('Product Name')['Profit'].sum()

top_sales_product = product_sales.idxmax()
top_sales_value = product_sales.max()

top_profit_product = product_profit.idxmax()
top_profit_value = product_profit.max()

print(f"Highest Sales Product: {top_sales_product} - {top_sales_value:}")
print(f"Highest Profit Product: {top_profit_product} - {top_profit_value:}")
```

Which product category (Furniture, Technology, Office Supplies) is the most profitable?

```python
category_profit = df.groupby('Category')['Profit'].sum()

most_profitable_category = category_profit.idxmax()
most_profit_value = category_profit.max()

print("Profit by Category: \n", category_profit)
print(f"\nMost Profitable Category: {most_profitable_category} - {most_profit_value:}")
```

# Data Visulation

```python
import matplotlib.pyplot as plt
category_sales = df.groupby('Category')['Total Sales'].sum()
category_sales.plot(kind='bar', color='skyblue')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.sort_values('Order Date', inplace=True)
daily_sales = df.groupby('Order Date')['Total Sales'].sum()
plt.figure(figsize=(10, 4))
plt.plot(daily_sales.index, daily_sales.values, color='green')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
region_sales = df.groupby('Region')['Total Sales'].sum()
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Region')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
plt.hist(df['Profit'], bins=50, color='purple')
plt.title('Profit Distribution')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(df['Total Sales'], df['Profit'], alpha=0.5, color='orange')
plt.title('Sales vs. Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

Heatmap

```python
import seaborn as sns
corr = df[['Total Sales', 'Quantity', 'Discount', 'Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Line Plot

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
daily_sales = df.groupby('Order Date')['Total Sales'].sum().reset_index()
sns.lineplot(data=daily_sales, x='Order Date', y='Total Sales', color='green')
plt.title('Sales Over Time')
plt.tight_layout()
plt.show()
```

Bar Plot

```python
sales_by_category = df.groupby('Category')['Total Sales'].sum().reset_index()
sns.barplot(data=sales_by_category, x='Category', y='Total Sales', palette='Blues_d')
plt.title('Total Sales by Category')
plt.tight_layout()
plt.show()
```

Histogram

```python
sns.histplot(df['Profit'], bins=50, kde=True, color='purple')
plt.title('Profit Distribution')
plt.tight_layout()
plt.show()
```
